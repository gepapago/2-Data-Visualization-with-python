# Το Νήμα της Ζωής
### Using only a small part of the World-Bank data, we will explore the relationship between GDP per capita and life expectancy in different countries.

* GDP per capita data available at http://data.worldbank.org/indicator/NY.GDP.PCAP.PP.CD.

* GDP has been reduced to equivalent purchasing power (Purchasing Power Parity, PPP).

```python
import pandas as pd
pd.set_option('display.max_colwidth', 25)
gdp_df = pd.read_csv('worldbank/gdp_per_capita_ppp/API_NY.GDP.PCAP.PP.CD_DS2_en_csv_v2_3469401.csv', 
                     skiprows=4)
```

```python
gdp_df
```

* Θα αφαιρέσουμε την τελευταία στήλη, η οποία δημιουργήθηκε επειδή κάθε γραμμή λήγει σε κόμμα.

```python
gdp_df.drop(gdp_df.columns[len(gdp_df.columns) - 1], inplace=True, axis=1)
```

```python
gdp_df
```

* Some of the lines do not correspond to specific countries, but to groups of countries.

* We can find out which lines these are by looking at the metadata the World Bank gives us.

```python
countries_df = pd.read_csv('worldbank/gdp_per_capita_ppp/Metadata_Country_API_NY.GDP.PCAP.PP.CD_DS2_en_csv_v2_3469401.csv')
countries_df
```

* Again we will remove the last column.

* Countries are on the lines that do not have a space in the `Region' column.

```python
countries_df.drop(countries_df.columns[len(countries_df.columns) - 1], inplace=True, axis=1)
countries_df = countries_df.loc[countries_df['Region'].notnull()]
countries_df
```

* Now we can join countries with GDP data to get actual countries with their GDP and geographic area.

```python
gdp_countries_df = pd.merge(gdp_df, 
                            countries_df[["Country Code", "Region"]], 
                            how="inner", 
                            on="Country Code")
```

```python
gdp_countries_df
```

* Next we will take the life expectancy data, which is available at <http://data.worldbank.org/indicator/SP.DYN.LE00.IN>.

```python
lex_df = pd.read_csv('worldbank/life_expectancy/API_SP.DYN.LE00.IN_DS2_en_csv_v2_3469289.csv', 
                     skiprows=4)
```

```python
lex_df
```

* Again we will remove the last column.

```python
lex_df.drop(lex_df.columns[len(lex_df.columns) - 1], inplace=True, axis=1)
```

```python
lex_df
```

* We will merge the GDP `DataFrame` with the Life Expectancy `DataFrame`.

* The union will be done on the `Country Code' column.

* We must notice that there are columns with the same name in both `DataFrame`s.

* Unless specified otherwise, pandas will add suffixes `_x` and `_y` to distinguish columns.

* To make it easier to distinguish the origin of the columns we will give our own endings.

```python
lex_gdp_countries_df = pd.merge(gdp_countries_df, 
                                lex_df.loc[:,'Country Code':], 
                                how="inner", 
                                on="Country Code",
                                suffixes=("_gdp", "_lex"))
```

```python
lex_gdp_countries_df
```

* We can now make a graph with the 2019 data (because this is the last year for which we have life expectancy data).

```python
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.figure(figsize=(10, 8))
_ = plt.scatter(x=lex_gdp_countries_df['2019_gdp'], y=lex_gdp_countries_df['2019_lex'])
```

* The image shows that the countries belong to two different categories:

 * A group with large differences in life expectancy and small differences in GDP per capita, which are at the left end of the figure.

 * A group with small differences in life expectancy and large differences in GDP per capita, roughly $25,000 and up.

```python
import matplotlib.patches as patches

plt.figure(figsize=(10, 8))
plt.scatter(x=lex_gdp_countries_df['2019_gdp'], y=lex_gdp_countries_df['2019_lex'])
rect1 = patches.Rectangle((0, 52), width=25000, height=30, color='blue', alpha=0.2)
rect2 = patches.Rectangle((25000, 72), width=120000, height=14, color='gold', alpha=0.2)
plt.gca().add_patch(rect1)
_ = plt.gca().add_patch(rect2)
```

* If we are interested in the mathematical relationship between income and life expectancy, it can be seen more clearly if we transform the $x$ axis so that the scale is logarithmic.

* Then we get a relation of the form:

 $$ y = a + b \log(x) $$

```python
plt.figure(figsize=(10, 8))
plt.scatter(x=lex_gdp_countries_df['2019_gdp'], y=lex_gdp_countries_df['2019_lex'])
plt.xscale('log')
```

* Let's also add population to the picture.

* Population is available from <http://data.worldbank.org/indicator/SP.POP.TOTL>.

* We remove the last column again.

```python
population_df = pd.read_csv('worldbank/population/API_SP.POP.TOTL_DS2_en_csv_v2_3603750.csv', 
                            skiprows=4)
population_df.drop(population_df.columns[len(population_df.columns) - 1], 
                   inplace=True, axis=1)
```

```python
population_df
```

* We will join the `DataFrame` that we have the GDP and life expectancy data with the `DataFrame` of the population.

* The union will be done again on the `Country Code' column.

```python
pop_lex_gdp_countries_df = pd.merge(lex_gdp_countries_df, 
                                    population_df, 
                                    how="inner", 
                                    on="Country Code")
```

```python
pop_lex_gdp_countries_df
```

* For documentation purposes we will rename the population columns to have the suffix `_pop`.

```python
pop_column_names = { c: f'{c}_pop' for c in population_df.columns if '1960' <= c <= '2020'}
pop_lex_gdp_countries_df.rename(columns=pop_column_names, inplace=True)
```

```python
pop_lex_gdp_countries_df
```

* To see what we can make considering the population as well.

```python
plt.figure(figsize=(10, 8))
_ = plt.scatter(x=pop_lex_gdp_countries_df['2019_gdp'], y=pop_lex_gdp_countries_df['2019_lex'], 
                s=pop_lex_gdp_countries_df['2019_pop']/(2*10**5),
                alpha=0.5)
```

* We can change the values ​​on the horizontal axis to be in thousands.

* Also, we can make it logarithmic again.

```python
from matplotlib.ticker import FuncFormatter

plt.figure(figsize=(10, 8))
plt.scatter(x=pop_lex_gdp_countries_df['2019_gdp'], 
            y=pop_lex_gdp_countries_df['2019_lex'], 
            s=pop_lex_gdp_countries_df['2019_pop']/(2*10**5),
            alpha=0.5)
plt.xscale("log")

formatter = FuncFormatter(lambda v, p: "%dK" % (v / 1000))

plt.gca().xaxis.set_major_formatter(formatter)
```

```python
from matplotlib.ticker import FuncFormatter

plt.figure(figsize=(10, 8))
plt.scatter(x=pop_lex_gdp_countries_df['2019_gdp'], 
            y=pop_lex_gdp_countries_df['2019_lex'], 
            s=pop_lex_gdp_countries_df['2019_pop']/(2*10**5),
            alpha=0.5)
plt.xscale("log")

formatter = FuncFormatter(lambda v, p: "%dK" % (v / 1000))

plt.gca().xaxis.set_major_formatter(formatter)
```

* To further improve our chart, we would like to group countries by region.

* To do this we will treat area as a categorical variable and assign a number to each category.

```python
region_cat = pop_lex_gdp_countries_df['Region'].sort_values().astype('category')
pop_lex_gdp_countries_df['Region Code'] = region_cat.cat.codes
```

```python
pop_lex_gdp_countries_df
```

* Thus, we can make a graph that also contains the geographical area of ​​each country.

```python
plt.style.use('seaborn-whitegrid')

matplotlib.rcParams.update({'font.size': 16})
plt.figure(figsize=(10, 8))
scatter = plt.scatter(x=pop_lex_gdp_countries_df['2019_gdp'], 
            y=pop_lex_gdp_countries_df['2019_lex'], 
            s=pop_lex_gdp_countries_df['2019_pop']/(2*10**5),
            c=pop_lex_gdp_countries_df['Region Code'],
            alpha=0.6,
            edgecolors='k',
            label=pop_lex_gdp_countries_df['Region'],
            cmap='Spectral_r')

handles, labels = scatter.legend_elements(prop="colors", num=len(region_cat.cat.categories), 
                                          size=10, mec='k', alpha=0.6)
labels = region_cat.cat.categories
plt.legend(handles, labels, loc="lower right")
plt.xscale("log")
formatter = FuncFormatter(lambda v, p: "%dK" % (v / 1000))
plt.gca().xaxis.set_major_formatter(formatter)
```
